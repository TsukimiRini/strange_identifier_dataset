```python
import os
import sys
from csv import DictReader
from loguru import logger

os.chdir("/Users/tannpopo/Projects/coder-ask-for-help/data_collection")

repo_csv_path = 'monitors4codegen/datasets/PragmaticCode/repos.csv'
with open(repo_csv_path) as f:
    repos = list(DictReader(f))

! mkdir -p data/dot_prompt_repos
os.chdir('data/dot_prompt_repos')

for repo in repos:
    logger.info(f"""Cloning {repo['repo']}""")
    repo_url = f"""https://github.com/{repo['repo']}"""
    repo_commit = repo['commit_id'][-7:]
    os.chdir("/Users/tannpopo/Projects/coder-ask-for-help/data_collection/data/dot_prompt_repos")
    if os.path.exists(repo['repo'].split('/')[-1]):
        logger.info(f"""{repo['repo']} already exists""")
        os.chdir(repo['repo'].split('/')[-1])
        os.system(f"""git checkout {repo_commit}""")
        continue
    os.system(f"""git clone {repo_url}""")
    if not os.path.exists(repo['repo'].split('/')[-1]):
        logger.info(f"""{repo['repo']} does not exist""")
        continue
    os.chdir(repo['repo'].split('/')[-1])
    os.system(f"""git checkout {repo_commit}""")
```
